```python
import pandas as pd  

data = {'Name': ['Alice', 'Bob', 'Charlie'], 
        'Age': [25, 30, 'Thirty-Five'],  
        'Gender': ['F', 'M', 'M']}  

df = pd.DataFrame(data)
df['Age'] = df['Age'].replace('Thirty-Five', 35)    #Replacing 'Age' to numeric and handle non-numeric values using '.replace'
average_age = df['Age'].mean()

print(f"Average Age: {average_age}")
```

```python
import numpy as np
from scipy.stats import zscore

data = np.array([1, 2, 3, 100, 5, 6, 7, 8, 9])
z_scores = zscore(data)

# Filter the data where the absolute Z-score is greater than 2
outliers = data[np.abs(z_scores) > 2]

print(outliers)
```

```python
import numpy as np
from scipy.stats import zscore

data = [1, 2, 3, 100, 5, 6, 7, 8, 9]
median = np.median(data)

data = [median if abs(zscore(x)) > 2 else x for x in data]
```

```python
import numpy as np
from scipy.stats import zscore

data = [1, 2, 3, 100, 5, 6, 7, 8, 9]

z_scores = zscore(data) 
median = np.median(data)

# Correct: Replace outliers with the median using proper Z-score comparison
data = [median if abs(z_scores[i]) > 2 else data[i] for i in range(len(data))]
print(data)
```
